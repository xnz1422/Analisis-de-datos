#Actividad 2: Corrección de Tipos de Datos y Normalización

Cargar el conjunto de datos.

```python
datos = "https://docs.google.com/spreadsheets/d/170GDFJKpZEaBbA48mII018sW6qBOGz2X4UlV4HNQaWo/export?format=xlsx"
```

```python
df_2 = pd.read_excel(datos)
```

```python
df_2.info()
df_2.head()
```

Corregir el tipo de datos de la columna de precio a un tipo numérico.

```python
df_2["Precio"] = df_2["Precio"].str.replace("$", "")
```

```python
df_2["Precio"] = df_2["Precio"].astype(float)
```

```python
df_2.info()
df_2.head()
```

Normalizar el nombre del producto para que todas las entradas estén en minúsculas y sin caracteres especiales.

```python
df_2["Producto"] = df_2["Producto"].str.lower()
```

```python
df_2.head()
```

```python
df_2["Producto"] = df_2["Producto"].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
df_2
```
