Utilizar el método isnull() para identificar las filas con datos faltantes y contar el número de valores nulos por columna.

```python
df_nulos = df.isnull()
df_nulos
```

```python
cant_nulos = df_nulos.sum()
cant_nulos
```

Usar el método duplicated() para identificar las filas duplicadas y contar cuántas filas duplicadas hay en total.

```python
df_duplicates = df.duplicated()
```

```python
df_duplicates
```

```python
cant_duplicates = df_duplicates.sum()
cant_duplicates
```

Crear un informe que incluya:

1.  La cantidad total de registros en el DataFrame.
2.  La cantidad total de valores nulos por columna.
3.  La cantidad de filas duplicadas.







```python
cant_registros = len(df)
cant_reg_nulos = cant_nulos
cant_reg_dup = cant_duplicates

print(f"""
Informe:

Se observan en el dataframe un total de {cant_registros} registros.

Tiene la siguiente cantidad de registros nulos por columna:\n\n{cant_reg_nulos}

Esta es la cantidad de filas duplicadas: {cant_duplicates}

""")
```

## Para el estudio estadistico
* Datos anomalos: No existen casos anomalos.
* Datos nulos: rellenar con la media de cada fila.

## Seguimiento de un paciente
* Datos anómalos: no se encontraron.
* Datos nulos: ignorar los datos.

## Detección de casos de contagio
* Datos anómalos: realizar un analisis más profundo de estos casos, examinar si los datos se repiten en la misma persona multipleces veces. Y si se repiten con multiples personas.
* Datos nulos: ignorar a menos que se conozca si la persona se contagio, en ese caso rellenar con la media.
